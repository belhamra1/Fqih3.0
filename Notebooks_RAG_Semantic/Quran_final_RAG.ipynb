### Semantic search using sentence transformers

````python
from sentence_transformers import SentenceTransformer, util

# Function that searches the corpus and prints the results
def search(inp_question):
    start_time = time.time()
    question_embedding = model1.encode(inp_question, convert_to_tensor=True)
    hits = util.semantic_search(question_embedding, embeddings)
    end_time = time.time()
    hits = hits[0]  #Get the hits for the first query

    print("Input question:", inp_question)
    print("Results (after {:.3f} seconds):".format(end_time-start_time))
    for i, hit in enumerate(hits[0:5], start=1):
      print(f"Score: {hit['score']:.3f}\nDocument {i}:\n{RAW_KNOWLEDGE_BASE[hit['corpus_id']]}")
      print("--------------------------")


search("لمن تعطى الصدقات")
````

### preparing the  docs  for Indexing

````python

import json
import torch



# Prepare documents, metadata, and IDs
documents = []
metadatas = []
ids = []

for index, entry in enumerate(data):  # Assuming data is a list of entries
    # Create document content
    document = (
        f"الاية: {entry.get('aya_no_chakl', '')}\n"
        f"التفسير: {entry.get('tafsir_aya', '')}"
    )
    documents.append(document)
    
    # Create metadata
    metadata = {
        "السورة": entry.get("sura_name", ""),
        "رقم الاية": entry.get("aya_index_sura", "")
    }
    metadatas.append(metadata)
    
    # Create unique ID
    ids.append(str(index + 1))

# Load saved embeddings


````

````python

# Convert the tensor to a list of lists (if needed)
embeddings_tensor = torch.load("/kaggle/working/embeddings_Quran_omar.pt", map_location=torch.device("cuda"), weights_only=True)

#type(embeddings_tensor)
embeddings_tensor.keys()
````

````python
embeddings_data = embeddings_tensor['embeddings']
print(type(embeddings_data))
````

### Create a collection and add the data to Chroma DB

````python
import chromadb
import numpy as np




# Initialize ChromaDB client
client = chromadb.PersistentClient(path="/kaggle/working/EMBEDD/")

# Create a collection for the Enron messages
collection_QURAN = client.create_collection(
    name="quran_embbed_STS",
    metadata={"hnsw:space": "cosine"} #dot product (ip) if the models use it
)

# Define the batch size
batch_size = 5000  # Adjust this size based on your needs and constraints

# Add embeddings to ChromaDB in batches
for i in range(0, len(documents), batch_size):
    end_index = min(i + batch_size, len(documents))
    batch_documents = documents[i:end_index]
    batch_embeddings = embeddings[i:end_index]
    batch_metadatas = metadatas[i:end_index]
    batch_ids = ids[i:end_index]

    collection_QURAN.add(
        documents=batch_documents,
        embeddings=batch_embeddings,
        metadatas=batch_metadatas,
        ids=batch_ids
    )

print("Data quran  successfully added to ChromaDB.")

````

````python
%%writefile app_chatbot.py
# Streamlit App

import streamlit as st
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
from typing import List, Tuple, Optional



from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores.utils import DistanceStrategy

import pandas as pd


from langchain_chroma import Chroma
import chromadb

import chromadb
import numpy as np

# Initialize ChromaDB client
client = chromadb.PersistentClient(path="/kaggle/working/EMBEDD/")

#collection_quran = client.get_collection("quran_embbed_Omarai_V2")




pd.set_option("display.max_colwidth", None)  # This will be helpful when visualizing retriever outputs

# Rerabking the retried docs : improve the process of retrieval
#RERANKER = RAGPretrainedModel.from_pretrained("colbert-ir/colbertv2.0")



embedding_model = HuggingFaceEmbeddings(model_name="Omartificial-Intelligence-Space/Arabic-Triplet-Matryoshka-V2")



vector_store_from_client = Chroma(
    client=client,
    collection_name="quran_embbed_Triplet",
    embedding_function=embedding_model
)





# Assuming your KNOWLEDGE_VECTOR_DATABASE and RERANKER are defined elsewhere and FAISS is available

# Define the answer_with_rag function here (from your code above)

def answer_with_rag(
    question: str,
    llm: pipeline,
    knowledge_index: Chroma,
    #reranker=None,
    num_retrieved_docs: int = 5,
    #num_docs_final: int = 5,
) -> Tuple[str, List[str]]:
    # Retrieving and optionally reranking documents
    relevant_docs = knowledge_index.similarity_search_with_score(query=question, k=num_retrieved_docs)
   # relevant_docs = [doc.page_content for doc in relevant_docs]

   

    context = "\nExtracted documents:\n"
    context += "".join([
            f"Document {str(i)}:::\nContent: {doc[0]}\nScore: {doc[1]:.3f}\n\n"
            for i, doc in enumerate(relevant_docs)
    ])

    final_prompt = RAG_PROMPT_TEMPLATE.format(question=question, context=context)

    answer = llm(final_prompt)[0]["generated_text"]
    return answer, relevant_docs

# Define FAISS-based retrieval here
# KNOWLEDGE_VECTOR_DATABASE = FAISS(...)

# Quantization config and loading the model
MODEL_LLM="Qwen/Qwen2.5-7B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_LLM,
    quantization_config=bnb_config
)
tokenizer = AutoTokenizer.from_pretrained(MODEL_LLM)

READER_LLM = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    do_sample=True,
    temperature=0.2,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=400
)

RAG_PROMPT_TEMPLATE = tokenizer.apply_chat_template(
    [
        {
            "role": "system",
            "content": """Using the information provided in the context, provide a precise and consistent answer to the question.
Respond only to the question asked, ensuring your response is concise and directly relevant.
Do not include Quranic verses from the context in your response.
Ensure the generated response reflects Quranic principles without directly quoting verses.
If the answer cannot be strictly deduced from the context provided, do not provide an answer.
Respond in Arabic.""",
        },
        {
            "role": "user",
            "content": """Context:
            {context}
            ---
            Now here is the question you need to answer.

            Question: {question}""",
        },
    ],
    tokenize=False,
    add_generation_prompt=True
)


#--------------------------------------------------

#streamlit app
import streamlit as st
st.title("RAG-based Quran Chatbot")

# Input: user query
user_query = st.chat_input("Enter your query - Arabic :")

if user_query:
    with st.chat_message("user"):
        st.write(user_query)


        # Perform the retrieval and generation
    with st.spinner(' Generating an answer...'):
            answer, relevant_docs = answer_with_rag(user_query, READER_LLM, vector_store_from_client)

        # Display the generated answer
    with st.chat_message("assistant"):
        st.write(answer)

    with st.expander("Retrieved Documents"):
        if relevant_docs:
            for i, (doc, score) in enumerate(relevant_docs):
                st.markdown(f"📄 Document {i + 1} (Score: {score:.3f})")
                st.markdown(f"**Similarity Score**: `{score:.3f}`")
                st.markdown(f"**Content**:\n\n{doc.page_content}")
                st.markdown(f"**Metadata**:\n\n```{doc.metadata}```")
        else:
            st.error("Please enter a query.")


````
